```python
import sympy as sy
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Problem 1
def prob1():
    """Return an expression for

        (2/5)e^(x^2 - y)cosh(x+y) + (3/7)log(xy + 1).

    Make sure that the fractions remain symbolic.
    """

    x, y = sy.symbols('x, y')
    return sy.Rational(2,5) * sy.exp(x**2 - y)*sy.cosh(x+y) + sy.Rational(3,7)*sy.log(x*y + 1)



# Problem 2
def prob2():
    """Compute and simplify the following expression.

        product_(i=1 to 5)[ sum_(j=i to 5)[j(sin(x) + cos(x))] ]
    """
    x, i, j = sy.symbols('x, i, j')
    sin = sy.sin(x)
    cos = sy.cos(x)

    return sy.simplify(sy.product(sy.summation(j * (sin + cos), (j, i, 5)), (i, 1, 5)))


# Problem 3
def prob3(N):
    """Define an expression for the Maclaurin series of e^x up to order N.
    Substitute in -y^2 for x to get a truncated Maclaurin series of e^(-y^2).
    Lambdify the resulting expression and plot the series on the domain
    y in [-3,3]. Plot e^(-y^2) over the same domain for comparison.
    """

    #Create expression
    x, y, n = sy.symbols('x, y, n')
    product = sy.lambdify(y, sy.summation(x**n / sy.factorial(n), (n, 0, N)).subs(x, -y**2), "numpy")
    #Plot actual function against Maclaurin series
    domain = np.linspace(-2, 2, 100)
    plt.plot(domain, product(domain), label="Maclaurin")
    plt.plot(domain, np.exp(-1 * domain**2), label="e^(-y^2)")
    plt.legend(loc="upper left")
    plt.show()




# Problem 4
def prob4():
    """The following equation represents a rose curve in cartesian coordinates.

    0 = 1 - [(x^2 + y^2)^(7/2) + 18x^5 y - 60x^3 y^3 + 18x y^5] / (x^2 + y^2)^3

    Construct an expression for the nonzero side of the equation and convert
    it to polar coordinates. Simplify the result, then solve it for r.
    Lambdify a solution and use it to plot x against y for theta in [0, 2pi].
    """
    #Create and simplify expression
    x, y, r, θ = sy.symbols('x, y, r, θ')
    expr = 1 - ((x**2 + y**2)**(7/2) + 18*x**5 * y - 60*x**3 * y**3 + 18*x*y**5) / (x**2 + y**2)**3
    expr = sy.simplify(expr.subs({x:r*sy.cos(θ), y:r*sy.sin(θ)}))
    #Find roots of the expression
    f = sy.lambdify(θ, sy.solve(expr, r)[0], "numpy")
    #Plot the rose curve
    domain = np.linspace(0, 2*np.pi, 400)
    plt.plot(f(domain)*np.cos(domain), f(domain)*np.sin(domain), label="flower")
    plt.legend(loc="upper left")
    plt.show()

# Problem 5
def prob5():
    """Calculate the eigenvalues and eigenvectors of the following matrix.

            [x-y,   x,   0]
        A = [  x, x-y,   x]
            [  0,   x, x-y]

    Returns:
        (dict): a dictionary mapping eigenvalues (as expressions) to the
            corresponding eigenvectors (as SymPy matrices).
    """
    #Create matrix A and Identity matrix
    x, y, λ = sy.symbols('x, y, λ')
    A = sy.Matrix([[x-y, x, 0], [x, x-y, x], [0, x, x-y]])
    I = sy.Matrix([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
    #Find eigenvalues
    expr = (A - λ*I).det()
    eigenvals = sy.solve(expr, λ)
    eigenvectors = []
    #Find eigenvalues
    for eigenvalue in eigenvals:
        eigenvectors.append((A - eigenvalue*I).nullspace())

    return {eigenvalue:eigenvectors[i] for i, eigenvalue in enumerate(eigenvals)}



# Problem 6
def prob6():
    """Consider the following polynomial.

        p(x) = 2*x^6 - 51*x^4 + 48*x^3 + 312*x^2 - 576*x - 100

    Plot the polynomial and its critical points. Determine which points are
    maxima and which are minima.

    Returns:
        (set): the local minima.
        (set): the local maxima.
    """
    #Find the critical values of p(x) and intialize lists
    x = sy.symbols('x')
    p = 2*x**6 - 51*x**4 + 48*x**3 + 312*x**2 - 576*x - 100
    plottable = sy.lambdify(x, p)
    criticals = sy.solve(sy.diff(p, x), x)
    minimums = []
    maximums = []
    unknowns = []
    #Fill lists with maximums and minimums
    for critical in criticals:
        if sy.lambdify(x, sy.diff(p, x, x))(critical) > 0:
            minimums.append(critical)
        elif sy.lambdify(x, sy.diff(p, x, x))(critical) < 0:
            maximums.append(critical)
        else:
            unknowns.append(critical)
    #Plot p(x) and the maximums and minimums
    domain = np.linspace(-5, 5, 100)
    plt.plot(domain, plottable(domain))
    plt.scatter(maximums, plottable(np.array(maximums)))
    plt.scatter(minimums, plottable(np.array(minimums)))
    plt.show()

    return set(minimums), set(maximums)


# Problem 7
def prob7():
    """Calculate the integral of f(x,y,z) = (x^2 + y^2 + z^2)^2 over the
    sphere of radius r. Lambdify the resulting expression and plot the integral
    value for r in [0,3]. Return the value of the integral when r = 2.

    Returns:
        (float): the integral of f over the sphere of radius 2.
    """
    p, t, phi, x, y, z, r = sy.symbols('ρ , θ, ϕ, x, y, z, r')
    #Create the matrix h(ρ, θ, ϕ)
    h_1 = p*sy.sin(phi)*sy.cos(t)
    h_2 = p*sy.sin(phi)*sy.sin(t)
    h_3 = p*sy.cos(phi)
    H = sy.Matrix([h_1, h_2, h_3])
    #Find the determinant of the Jacobian of H
    J = sy.simplify(H.jacobian([p, t, phi]))
    det = -1*J.det()
    #Simplify f(h(ρ, θ, ϕ)) and integrate the desired expression
    f = sy.simplify((h_1**2 + h_2**2 + h_3**2)**2)
    integrated = sy.integrate(f*det, (p, 0, r), (t, 0, 2*sy.pi), (phi, 0, sy.pi))
    #plot the integral for varying values of r and return the value of the integral at r = 2
    lambdified = sy.lambdify(r, integrated)
    domain = np.linspace(0, 3, 100)
    plt.plot(domain, lambdified(domain))
    plt.show()
    return lambdified(2)
```
