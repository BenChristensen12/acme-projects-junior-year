```python
# Problem 1
def arithmagic():
    """Perform arithmagic. Raise errors if input is invalid.
    """
    step_1 = input("Enter a 3-digit number where the first and last "
                                           "digits differ by 2 or more: ")
    if int(step_1) < 100 or int(step_1) > 999:
        raise ValueError("You did not enter a 3-digit number")
    if abs(int(step_1[0]) - int(step_1[2])) < 2:
        raise ValueError("First and last digits differ by less than two")

    step_2 = input("Enter the reverse of the first number, obtained "
                                              "by reading it backwards: ")
    if len(step_2) != len(step_1):
        raise ValueError("That is not the reverse of the first number")
    for i in range(3):
        if step_1[i] != step_2[2 - i]:
            raise ValueError("That is not the reverse of the first number")

    step_3 = input("Enter the positive difference of these numbers: ")
    if int(step_3) != abs(int(step_1) - int(step_2)):
        raise ValueError("That is not the positive difference of the first two numbers")

    step_4 = input("Enter the reverse of the previous result: ")
    if len(step_4) != len(step_3):
        raise ValueError("That is not the reverse of the last number")
    for i in range(3):
        if step_3[i] != step_4[2 - i]:
            raise ValueError("That is not the reverse of the last number")
    print(str(step_3), "+", str(step_4), "= 1089 (ta-da!)")

# Problem 2
def random_walk(max_iters=1e12):
    """Perform a random walk. Abort when the keyboard interrupt key is struck.
    """
    walk = 0
    directions = [1, -1]
    try:
        for i in range(int(max_iters)):
            walk += choice(directions)
    except KeyboardInterrupt:
        print("Process interrupted at iteration", i)
    else:
        print("Process completed.")
    finally:
        return walk


# Problems 3 and 4: Write a 'ContentFilter' class.
class ContentFilter:
    """Read a file, store its contents in a string that can be accessed as
        an attribute, securely close the file, continue prompting for input
        whenever input is invalid


    Attributes:
        name (str): Name of the file to be read
        contents (str): String of contents of the file
        acceptable (list): List of acceptable keys to be used in function
            parameters for problem 4.
    """

    def __init__(self, name):
        """Set the name of the file to be read.

        Parameters:
            name (str): Name of the file to be read
        """
        file_error = True
        while file_error == True:
            try:
                file_error = False
                thisfile = open(name)
            except (TypeError, FileNotFoundError, OSError) as e:
                name = input("Please enter a valid file name: ")
                file_error = True
            else:
                self.contents = thisfile.read()
                thisfile.close()

        self.name = name
        self.acceptable = ['w', 'x', 'a']

    def test_mode(self, mode):
        """Test if the mode entered for each function is an acceptable mode
        """
        bad_mode = True
        for i in range(3):
            if self.acceptable[i] == mode:
                bad_mode = False
        if bad_mode:
            raise ValueError("Enter a valid file access mode: 'w', 'x', or 'a'")

    def uniform(self, to_file, mode='w', case="upper"):
        """Write uniformly upper or lower case content from the self file to
            a desired file.
        """
        ContentFilter.test_mode(self, mode)
        if case != "upper" and case != "lower":
            raise ValueError('Case must be "upper" or "lower"')
        with open(to_file, mode) as outfile:
            for i in range(len(self.contents)):
                if case == "upper":
                    outfile.write(self.contents[i].upper())
                elif case == "lower":
                    outfile.write(self.contents[i].lower())

    def reverse(self, to_file, mode='w', unit="line"):
        """Write the file reversed either by line or by word order
        """
        ContentFilter.test_mode(self, mode)
        if unit != "line" and unit != "word":
            raise ValueError('unit must be "line" or "word"')
        with open(to_file, mode) as outfile:
            if unit == "line":
                for j in range(len(self.contents.split('\n'))):
                    outfile.write(self.contents.split('\n')[-j - 1])
                    if j < len(self.contents.split('\n')):
                        outfile.write("\n")
            if unit == "word":
                for i in range(len(self.contents.split('\n'))):
                    #Next line is temp testing
                    #print([line[::-1] for line in self.contents.split('\n')][i])
                    outfile.write([line[::-1] for line in self.contents.split('\n')][i])
                    if i < len(self.contents.split('\n')) - 1:
                        outfile.write("\n")

    def transpose(self, to_file, mode='w'):
        """Transpose the contents of a file and write to a new file
        """
        ContentFilter.test_mode(self, mode)
        with open(to_file, mode) as outfile:
            line = dict()
            for i in range(len(self.contents.split('\n')) - 1):
                line[i] = self.contents.split('\n')[i]
            for j in range(len(line[0].split())):
                for k in range(len(line)):
                    outfile.write(line[k].split()[j])
                    if k < len(line) - 1:
                        outfile.write(' ')
                if j < (len(line[0].split())):
                    outfile.write('\n')

    def __str__(self):
        string = "Source file: \t\t" + self.name + "\n"
        string += "Total characters: \t" + str(len(self.contents)) + "\n"
        string += "Alphabetic characters: \t" + str(sum([s.isalpha() for s in self.contents])) + "\n"
        string += "Numerical characters: \t" + str(sum([s.isdigit() for s in self.contents])) + "\n"
        string += "Whitespace characters: \t" + str(sum([s.isspace() for s in self.contents])) + "\n"
        string += "Number of lines:\t" + str(len(self.contents.split('\n')))
        return string
```
